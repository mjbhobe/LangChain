## Query PDFs with Google Gemini LLM

```python
import pathlib
import textwrap
import os
from dotenv import load_dotenv, find_dotenv

import google.generativeai as genai

# Used to securely store your API key
# from google.colab import userdata

from IPython.display import display
from IPython.display import Markdown
```

```python
# load env variables from .env file
_ = load_dotenv(find_dotenv())  # read local .env file

# required to access Gemini API
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])
```

```python
def to_markdown(text):
    """function to convert response from model to correct format"""
    text = text.replace("â€¢", "  *")
    return Markdown(textwrap.indent(text, "> ", predicate=lambda _: True))
```

Let's make some API calls directly to Gemini Pro model. This can be done as shown in the cells below.

```python
# create an instance of the Gemini Pro LLM
llm = genai.GenerativeModel("gemini-pro")
```

```python
response = llm.generate_content(
    "Tell me the important capabilities of Google's Gemini Pro LLM"
)
to_markdown(response.text)
```

```python
# cheeky question ;)
response = llm.generate_content(
    "Is Google's Gemini Pro better than Open AI's Chat-GPT 4 LLM?"
)
to_markdown(response.text)
```

## Invoking Gemini API via Langchain

```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
chat_model = ChatGoogleGenerativeAI(model="gemini-pro")
```

```python
result = chat_model.invoke(
    "Can you suggest a good Linux powered laptop configuration for Deep Learning model development?"
)
to_markdown(result.content)
```

### Asking Gemini about images
When asking Gemini about images, be sure to use the `gemini-pro-vision` model instead of the `gemini-pro` model. The latter is oriented to textual generation tasks, whereas the former can handle images as well.

Here we will provide a URL of an image and ask Gemini to describe it.

```python
import pathlib, os

image_url = "https://picsum.photos/seed/picsum/200/300"
local_file_path = pathlib.Path(os.getcwd()) / "Muffin_in_a_bag.jpg"
```

```python
# here is the image in question
from PIL import Image
from IPython.display import display

# /home/mjbhobe/code/git-projects/LangChain/gemini/Muffin_in_a_bag.jpg
img = Image.open(local_file_path)
display(img)
```

```python
from langchain_core.messages import HumanMessage
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-pro-vision")
# example
message = HumanMessage(
    content=[
        {
            "type": "text",
            "text": "Describe what you see in this image?",
        },  # You can optionally provide text parts
        {"type": "image_url", "image_url": str(local_file_path)},
    ]
)
chat_model.invoke([message])
```
