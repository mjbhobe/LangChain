## Using Anthropic Claude

```python
import os
from dotenv import load_dotenv

# load all environment variable (API keys) from .env file
load_dotenv()

CLAUDE_MODEL = "claude-3-5-sonnet-20240620"
```

```python
# example of using Claude API
import anthropic

client = anthropic.Anthropic(
    # defaults to os.environ.get("ANTHROPIC_API_KEY")
    api_key=os.getenv("ANTHROPIC_API_KEY"),
)
```

```python
user_message = "Can you explain how I can use an Agent in my LLM based application"

message = client.messages.create(
    model=CLAUDE_MODEL,
    max_tokens=1024,
    temperature=0.3,
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": user_message,
                }
            ],
        }
    ],
)
```

```python
# create the model
from langchain_anthropic import ChatAnthropic


llm = ChatAnthropic(model=CLAUDE_MODEL, temperature=0.6, max_tokens=1024)
```

```python
llm.invoke("What is Generative AI?")
```
